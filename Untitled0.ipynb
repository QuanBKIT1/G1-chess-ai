```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Phân lớp dựa trên feature median_income

# Tạo 1 seri phân lớp từ 1,2,3,4,5 cho median_income
housing['income_cat'] = np.ceil(housing['median_income']/1.5)
housing['income_cat'].where(housing['income_cat']<5, other=5, inplace=True)

# This can help to improve the accuracy and generalization of the model.
from sklearn.model_selection import StratifiedShuffleSplit

spliter = StratifiedShuffleSplit(n_splits=1,test_size=0.2, random_state=42)

for train_index, test_index in spliter.split(housing,housing['income_cat']):
  strait_train_set = housing.loc[train_index]
  strait_test_set = housing.loc[test_index]

strait_train_set.drop('income_cat',axis=1, inplace=True)
strait_test_set.drop('income_cat',axis=1, inplace=True)
```

```python
housing = strait_train_set.copy()
# Plot 
housing.plot(kind='scatter',x='longitude', y='latitude',alpha=0.4,s=housing['housing_median_age']/100,c='median_house_value'\
             ,cmap = plt.get_cmap('jet'),colorbar =True, label= 'population',)
plt.legend()
```
